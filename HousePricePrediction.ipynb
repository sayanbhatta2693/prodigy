```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.columns
```

```python
df=df[['SalePrice','LotArea','BedroomAbvGr','BsmtFinSF2','BsmtFinSF1','MasVnrArea','GrLivArea','LowQualFinSF','2ndFlrSF','1stFlrSF','TotalBsmtSF','BsmtUnfSF','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','FullBath','HalfBath']]
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
X_train=df.drop('SalePrice',axis=1)
y_train=df['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
df['SalePrice'].head()
```

```python
y_predict=model.predict(df.iloc[:5,1:])
y_predict
```

```python
model.score(X_train,y_train)
```
